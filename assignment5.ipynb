```python
import numpy as np

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

sum_all_elements = np.sum(gfg)

sum_row_wise = np.sum(gfg, axis=1)

sum_column_wise = np.sum(gfg, axis=0)

print("Sum of all elements:", sum_all_elements)
print("Sum of all elements row-wise:", sum_row_wise)
print("Sum of all elements column-wise:", sum_column_wise)
```

```python
import numpy as np

array = np.array([10, 52, 62, 16, 16, 54, 453])
print(f"Original array : {array}")
array.sort()

sorted_indices = np.argsort(array)

smallest_elements = array[:4]

largest_elements = array[-5:]

print(f"Sorted Array : {array}")
print(f"Sorted Indices {sorted_indices}")
print (f"Smallest elements : {smallest_elements}")
print (f"Largest elements : {largest_elements}")

array2 = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array2[array2 % 1 == 0]
print("Integer Elements Only:", integer_elements)

float_elements = array2[array2 % 1 != 0]
print("Float Elements Only:", float_elements)
```

```python
import numpy as np

# Step a: Generate unique sales dataset
initials = "AB"  # Replace with your initials
X = sum(ord(char) for char in initials)  # Sum of ASCII values
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales Data:", sales)

# Step b: Compute personalized tax rate
tax_rate = ((X % 5) + 5) / 100
sales_after_tax = sales * (1 - tax_rate)
print("Sales after Tax:", sales_after_tax)

# Step c: Adjust sales based on discount
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("Sales after Discount:", discounted_sales)

# Step d: Expand sales data for multiple weeks
weeks = 3
weekly_sales = np.tile(sales, (weeks, 1)).astype(float)

# Increase sales by 2% per week
for week in range(1, weeks):
    weekly_sales[week] *= 1.02  # Increase sales for each week

print("Weekly Sales Data:")
print(weekly_sales)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y_square = x**2
y_sin = np.sin(x)
y_exp = np.exp(x)
y_log = np.log(np.abs(x) + 1)

plt.figure(figsize=(10, 6))

plt.plot(x, y_square, label='y = x^2', linestyle='--')
plt.plot(x, y_sin, label='y = sin(x)')
plt.plot(x, y_exp, label='y = e^x', linestyle='-.')
plt.plot(x, y_log, label='y = log(|x| + 1)', linestyle=':')

plt.title("Function Plots")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid(True)

plt.show()
```
